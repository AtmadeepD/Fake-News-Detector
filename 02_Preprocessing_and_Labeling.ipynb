```python
import pandas as pd

# Load the cleaned dataset (if you haven't already in this session)
cleaned_liar_dataset = pd.read_csv('/kaggle/input/dataset/cleaned_liar_dataset.csv')

# Original labels in LIAR:
# ['false', 'half-true', 'mostly-true', 'true', 'pants-fire', 'barely-true']

# Map to binary
label_map = {
    'true': 'REAL',
    'mostly-true': 'REAL',
    'half-true': 'REAL',
    'false': 'FAKE',
    'pants-fire': 'FAKE',
    'barely-true': 'FAKE'
}

# Apply the mapping (if you haven't already)
cleaned_liar_dataset['label_binary'] = cleaned_liar_dataset['label'].map(label_map)

# Print the value counts of the 'label_binary' column
print(cleaned_liar_dataset['label_binary'].value_counts())
```
